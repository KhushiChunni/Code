# LINEAR REGRESSION


Importing the data set into pandas DataFrame


```python
import pandas as pd
import numpy as np
import sklearn
bos1=pd.read_csv('BostonHousing.csv')
bos1.head(10)
```

Processiong the data:Removing NaN values


```python
bos1.isna().sum()
```

Splitting model data with 70% for training

```python
from sklearn.model_selection import train_test_split

x=np.array(bos1.iloc[:,0:13])
y=np.array(bos1["medv"])
```

```python
#trsting dataset is of 30% of entire dataset
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=5)
```

Using Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
```

```python
#load our model
lr=LinearRegression()
```

```python
#Train the model using testing data
lr.fit(x_train,y_train)
```

```python
#Predict the testing data so that we can later evaluate the model
pred_lr=lr.predict(x_test)
```

Model evaluation


```python
#error for Linear Regression
mse_lr=sklearn.metrics.mean_squared_error(y_test,pred_lr,squared=False)
print("Error for linear regression ={}".format(mse_lr))
```

# Naive Bayes Theorem


Importing required libraries


```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

Importing training and testing dataset

```python
salary_train=pd.read_csv("SalaryData_Train(1).csv")
salary_test=pd.read_csv("SalaryData_Test(1).csv")
string_colums=["workclass",	"education","maritalstatus","occupation","relationship","race","sex",	"native"]
salary_train.head()
```

Using label encoder -Categorial to numerical


```python
from sklearn import preprocessing
number=preprocessing.LabelEncoder()
for i in string_colums:
  salary_train[i]=number.fit_transform(salary_train[i])
  salary_test[i]=number.fit_transform(salary_test[i])

```

Assign the encoded variables for training and testing


```python
colnames=salary_train.columns
len(colnames[0:13])
trainX=salary_train[colnames[0:13]]
trainY=salary_train[colnames[13]]
testX=salary_test[colnames[0:13]]
testY=salary_test[colnames[13]]
```

Gaussian naive bbayes classifier for model fitting


```python
sgnb=GaussianNB()
```

Prediction for test data

```python
spred_gnb=sgnb.fit(trainX,trainY).predict(testX)
```

Confusion Matrix


```python
from sklearn.metrics import confusion_matrix
cm_gnb=confusion_matrix(testY,spred_gnb)
print(cm_gnb)
```
