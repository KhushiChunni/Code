# Normalization


Normalization: To have mean and standard deviation of 0 and 1 respectively

```python
data=np.array([1,2,3,4,5])
```

```python
#Calculating mean and standard deviation
mean=np.mean(data)
std_dev=np.std(data)
```

```python
#Normalize the data

normalized_data=(data-mean)/std_dev
print("Normalized data:",normalized_data)
```

Statstical operations


```python
arr=np.array([1,2,3,4,5])
#Mean

print("Mean=",np.mean(arr))

#Median
print("median=",np.median(arr))

#Standard deviation

print("Std_dev=",np.std(arr))

#variance

print("Variance=",np.var(arr))

```

Logical Operations

```python
arr=np.array([1,2,3,4,5])

#Logical operation

print(arr[arr>2])
print(arr[(arr>2) & (arr<5)])
```

# Data Visualization With Matplotlib

Matplotlib is a powerful plotting library for Python that enables the creation of static, animated, and interactive visualizations. It is widely used for data visualization in data science and analytics. In this lesson, we will cover the basics of Matplotlib, including creating various types of plots and customizing them.

```python
import matplotlib.pyplot as plt
```

Line plot

```python
x=[1,2,3,4,5]
y=[1,4,7,19,45]

##create a line plot
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Basic line plot')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,7,19,45]

#customized line plot
plt.plot(x,y,color='red',linestyle='--')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Basic line plot')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,7,19,45]

#customized line plot with linestyle
plt.plot(x,y,color='red',linestyle='-.')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Basic line plot')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,7,19,45]

#customized line plot with marker
plt.plot(x,y,color='red',linestyle='--',marker='*')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Basic line plot')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,7,19,45]

#customized line plot with linewidth
plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=3)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Basic line plot')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,7,19,45]

#customized line plot with grid
plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=3)
plt.grid(True)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Basic line plot')
plt.show()
```

Multiple plots

```python
#sample data

x=[1,2,3,4,5]
y1=[1,4,7,19,45]
y2=[54,6,8,9,10]

plt.figure(figsize=(5,5))

plt.subplot(1,2,1)#1 row and 2 column
plt.plot(x,y1,color='pink')
plt.title("Plot 1")
```

```python
x=[1,2,3,4,5]
y1=[1,4,7,19,45]
y2=[54,6,8,9,10]

plt.figure(figsize=(5,2))

plt.subplot(1,2,1)#1 row and 2 column
plt.plot(x,y1,color='pink')
plt.title("Plot 1")

plt.subplot(1,2,2)#1 row and 2 column and plot on 2
plt.plot(x,y2,color='green')
plt.title("Plot 2")
```

```python
x=[1,2,3,4,5]
y1=[1,4,7,19,45]
y2=[54,6,8,9,10]

plt.figure(figsize=(5,5))

plt.subplot(2,2,1)#1 row and 2 column
plt.plot(x,x,color='pink')
plt.title("Plot 1")

plt.subplot(2,2,2)#1 row and 2 column
plt.plot(x,y2,color='green')
plt.title("Plot 2")
plt.subplot(2,2,3)#1 row and 2 column
plt.plot(x,y1,color='blue')
plt.title("Plot 3")
plt.subplot(2,2,4)#1 row and 2 column
plt.plot(y1,y2,color='yellow')
plt.title("Plot 4")
```

# Data visualisation using Seaborn

#### Data Visualization With Seaborn
Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code. In this lesson, we will cover the basics of Seaborn, including creating various types of plots and customizing them.

```python
import seaborn as sns
```

Basic plotting with seaborn

```python
tips=sns.load_dataset('tips')
tips
```

```python
import matplotlib.pyplot as plt
```

```python
#scatter plot
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter plot of total bill vs tip ")
```

```python
##Line plot

sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("line plot of total bill by size")
plt.show()
```

```python
#bar plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title("Bar plot of tatal bill vs day")
plt.show()
```

```python
#Box plot(for outliers)
sns.boxplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
##Voilon plot

sns.violinplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
##histogram
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
#Kde plot
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
#pair plot
sns.pairplot(tips)
```

```python
#heatmap
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
