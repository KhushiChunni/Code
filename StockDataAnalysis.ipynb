 1. Using Python, fetch historical stock data for any 5 stocks from the Nifty 50 index using the 'yfinance' library.Use a time range of the last 6 months.


```python
#Required libraries:

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# List of 5 stocks from Nifty 50
stocks = ['HDFCLIFE.NS', 'CIPLA.NS', 'WIPRO.NS', 'RELIANCE.NS', 'NESTLEIND.NS']
```

```python
# Fetching data for the last 6 months
data = yf.download(stocks,start="2024-01-02",end="2024-07-02",group_by='ticker')
```

```python
#First 5 rows of data
data.head()
```

```python
data.shape
```

2. Perform the following analyses on the data:

2-
   a. Calculate the daily percentage change for each stock.

```python
#Calculating the daily  percent change for each stock
percentage_change = {}
for stock in stocks:
    percentage_change[stock] = data[stock]['Adj Close'].pct_change() * 100

```

```python
percentage_change_df = pd.DataFrame(percentage_change)

percentage_change_df.head(10)
```

2-b. Plot the closing prices and daily percentage changes for the 5 stocks.

```python
#Plotting the closing price and daily percent changes
figure,axs = plt.subplots(5,2,figsize=(20, 20))
for i,stock in enumerate(stocks):
    # Closing prices
    axs[i, 0].plot(data[stock].index, data[stock]['Adj Close'],color='red',label=f'{stock} Closing Price')
    axs[i, 0].set_title(f'{stock} CLOSING PRICES')
    axs[i, 0].legend()
    # Daily percentage change
    axs[i, 1].plot(percentage_change_df.index, percentage_change_df[stock],color='teal',label=f'{stock} DAILY PERCENTAGE CHANGE')
    axs[i, 1].set_title(f'{stock} DAILY PERCENTAGE CHANGE')
    axs[i, 1].legend()


plt.show()


```

2-c. Calculate the moving average (20 days) for each stock and plot it against the stock's closing price.

```python
#Calculating 20-day moving average for each stock
moving_average_20 = {}
for stock in stocks:
    moving_average_20[stock] = data[stock]['Adj Close'].rolling(window=20).mean()


moving_average_20_df = pd.DataFrame(moving_average_20)
print(moving_average_20_df.shape)
print(moving_average_20_df.tail())
```

```python
# Plotting 20-day Moving average against closing prices
fig, axs = plt.subplots(5,1,figsize=(10, 20))

for i, stock in enumerate(stocks):
    axs[i].plot(data[stock].index, data[stock]['Adj Close'],color='red',label=f'{stock} CLOSING PRICE')
    axs[i].plot(moving_average_20_df.index, moving_average_20_df[stock],color='teal',label=f'{stock} 20-DAY MOVING AVERAGE')
    axs[i].set_title(f'{stock} CLOSING PRICE VS 20-DAY MOVING AVERAGE')
    axs[i].legend()

plt.show()


```

1. Develop a simple moving average crossover strategy for one of the 5 stocks. The strategy should involve a short-term moving average (e.g., 50-day) and a long-term moving average (e.g., 200-day)


```python
import numpy as np
```

```python
df=data['WIPRO.NS']
df.tail()
```

```python
df.loc[:,'SMA_20'] = df['Adj Close'].rolling(window=20).mean()
df.loc[:,'SMA_80'] = df['Adj Close'].rolling(window=80).mean()


df.head()
```

```python
# Define the signals: Buy when short SMA crosses above long SMA, and Sell when it crosses below
df.loc[:,'Signal'] = 0  # Default to no position
df.loc[:,'Signal'] = np.where(df['SMA_20'] > df['SMA_80'], 1, 0)  # Buy signal
df.loc[:,'Signal'] = np.where(df['SMA_20'] < df['SMA_80'], -1, df['Signal'])  # Sell signal

```

```python
# Shift the signals to generate the actual trading orders (so that buy/sell happens the next day)
df.loc[:,'Position'] = df['Signal'].shift()
```

```python
# Plotting the strategy with stock price and moving averages
plt.figure(figsize=(14, 7))
plt.plot(df['Adj Close'], label='WIPRO Adj Close', color='black', alpha=0.5)
plt.plot(df['SMA_20'], label='20-day SMA', color='blue', alpha=0.8)
plt.plot(df['SMA_80'], label='110-day SMA', color='red', alpha=0.8)

```

```python

print(df[['Adj Close', 'SMA_20', 'SMA_80', 'Signal']].tail())
```

```python
# Function to calculate maximum drawdown
def calculate_max_drawdown(data):
    roll_max = data['Adj Close'].cummax()
    daily_drawdown = data['Adj Close'] / roll_max - 1.0
    max_drawdown = daily_drawdown.cummin()
    return max_drawdown.min()
```

```python
# Calculating maximum drawdown for each stock
max_drawdowns = {}
for stock in stocks:
    max_drawdown = calculate_max_drawdown(data[stock])
    max_drawdowns[stock] = max_drawdown

```

```python
# Printing the results
for stock, drawdown in max_drawdowns.items():
    print(f"{stock} Maximum Drawdown: {drawdown:.2%}")
```

 2. Implementing Stop-Loss Mechanism

In Task 2, we developed a moving average crossover strategy for WIPRO. Now, we will introduce a stop-loss of 5%. This means if the stock price drops by 5% from the buy price, we trigger a sell action.

Here’s how you can implement the stop-loss:

```python
# Implementing Stop-Loss Mechanism for the Moving Average Crossover Strategy
stop_loss_percentage = 0.05  # 5% stop loss

```

```python
# Re-initializing portfolio variables
portfolio_value = []
cash = initial_investment
shares = 0
stop_loss_price = None  # Track stop-loss price
```

```python
# Iterating over each trading day to apply the strategy with stop-L
for i, row in df.iterrows():
    # Update stop-loss price based on buy price
    if shares > 0:
        stop_loss_price = max(stop_loss_price, row['Adj Close'] * (1 - stop_loss_percentage))
```

```python
 # Buy when crossover signal is generated

if row['Signal']==1 and cash > 0:
        shares = cash / row['Adj Close']
        cash = 0
        stop_loss_price = row['Adj Close'] * (1 - stop_loss_percentage)
```

```python
# Set stop-loss price
print(f"Buy on {i}: Price = {row['Adj Close']:.2f}")
```

```python
# Sell when crossover sell signal is generated or stop-loss is hit
if (row['Signal'] == -1 or (stop_loss_price is not None and row['Adj Close'] <= stop_loss_price)) and shares > 0:
    cash = shares * row['Adj Close']
    shares = 0
    sell_reason = "stop-loss" if stop_loss_price is not None and row['Adj Close'] <= stop_loss_price else "crossover"
    stop_loss_price = None  # Reset stop-loss price after determining the reason
    print(f"Sell on {i}: Price = {row['Adj Close']:.2f}, Reason: {sell_reason}")
```

```python
# Final portfolio value with stop-loss
final_portfolio_value_with_stop_loss = cash if shares == 0 else shares * df['Adj Close'].iloc[-1]
print(f"Final portfolio value with stop-loss: {final_portfolio_value_with_stop_loss:.2f}")

```
